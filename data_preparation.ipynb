```python
# Importing libraries and modules
from data_info import DataFrameInfo, DataFrameTransform, Plotter
import pandas as pd
```

```python
# function for loading data in csv format and returning pandas dataframe
def csv_dataframe(csv_file):
    df = pd.read_csv(csv_file)
    return df
```

```python
# creating the dataframe
loan_df = csv_dataframe('loan_payments.csv')
# creating two instances of two classes in data_info module
loan_info = DataFrameInfo(loan_df)
loan_tf = DataFrameTransform(loan_df)
```

```python
# stripping months and converting term column values into numeric ones
loan_tf.term_column_type('months')
```

```python
loan_tf.column_to_categorical(['grade', 'sub_grade', 'home_ownership',
                              'verification_status', 'loan_status', 'payment_plan', 'purpose'])
```

```python
# converting object columns to datetime ones
loan_tf.object_to_datetime('issue_date', '%b-%Y')
loan_tf.object_to_datetime('earliest_credit_line', '%b-%Y')
loan_tf.object_to_datetime('last_payment_date', '%b-%Y')
loan_tf.object_to_datetime('next_payment_date', '%b-%Y')
loan_tf.object_to_datetime('last_credit_pull_date', '%b-%Y')
```

## Skewness

The best way to represent skewness is through histogram or mathematically using pandas skew method. We'll use only visualisation tools since we have category variables on which the skew method cannot be applied.

#### Print Skewness for all columns

```python
print(loan_df.hist(figsize=(20,15)))
```

#### Positive skews

From the histograms we can conclude that the following variables have positive skew:
1. loan_amount
2. funded_amount
3. funded_amount_inv
4. instalment
5. open_accounts
6. total accounts
7. out_prncp
8. out_prncp_inv
9. total_payment
10. total_payment_inv
11. total_rec_prncp
12. total_rec_int
13. last_payment_amount



Because all these values are numerical, let's try to confirm the assumptions using pandas skew method

```python
loan_df[['loan_amount', 'funded_amount', 'funded_amount_inv', 'instalment', 'open_accounts', 'total_accounts', 'out_prncp', 'out_prncp_inv', 'total_payment', 'total_payment_inv', 'total_rec_prncp', 'total_rec_int', 'last_payment_amount']].skew()
```

***All columns above have values above 0.5 which suggest they are moderetaly and some highly skewed!***

#### Negative skews

The following columns suggest negative skewness from the histograms:
1. issue_date
2. earliest_credit_line
3. last_payment_date
4. last_credit_pull_date

***All these columns are of datetime data type and therefore I have decided not to normalise/transform the values of these values.***

**Let's try Yeo_Johnson transform for resolving positive skewness identified earlier. The reason for choosing it is that some of the variables have negative values and Box-Cox works only for positive values**

```python
loan_tf.yj_transform('loan_amount')
loan_tf.yj_transform('funded_amount')
loan_tf.yj_transform('funded_amount_inv')
loan_tf.yj_transform('instalment')
loan_tf.yj_transform('open_accounts')
loan_tf.yj_transform('total_accounts')
loan_tf.yj_transform('out_prncp')
loan_tf.yj_transform('out_prncp_inv')
loan_tf.yj_transform('total_payment')
loan_tf.yj_transform('total_payment_inv')
loan_tf.yj_transform('total_rec_prncp')
loan_tf.yj_transform('total_rec_int')
loan_tf.yj_transform('last_payment_amount')
```

```python
loan_df.info()
```

```python
# let's print the histogram of the transformed variables again
loan_df[['loan_amount', 'funded_amount', 'funded_amount_inv', 'instalment',
         'open_accounts', 'total_accounts', 'out_prncp', 'out_prncp_inv', 
        'total_payment', 'total_payment_inv', 'total_rec_prncp',
        'total_rec_int', 'last_payment_amount']].hist(figsize=(20,15))
```

**All the values are now near zero apart from out_prncp and out_prncp_inv. However, their skew value is much less than it was so for now I'll leave it as it is.**

## Colinearity

```python
# first, let's get numeric columns only
numeric_columns = loan_df.select_dtypes(include=['number'])
```

```python
import seaborn as sns
```

```python
# pd.set_option('display.max_columns', None)
# see the correlation matrix
sns.heatmap(numeric_columns.corr(), cmap='coolwarm')
```

The above table proves the following strong correlations:
1. total_payment and total_payment_inv with total_rec_prncp
2. loan_amount, funded_amount, funded_amount_inv with installment
3. funded_amount, funded_amount_inv with loan_amount
4. member_id with id

There are also strong correlations between payments and amounts and inv columns but that is expected since inv columns are proportional to their appropriate columns. 
Every column that has a correlation with other column above 0.90 or -0.90 has been considered as strong correlation and therefore the following columns will be dropped from the dataset:
1. id
2. total_rec_prncp
Installment is expected to be strongly correlated with the loan_amount and funded_amount and would not consider it at this stage as an issue and the same applies to the loan_amount and funded_amount correlation. Strong correlation in such cases can somtimes be helpful in prediction of a third variable.

Additionaly, unnamed column looks like a columns we have no information about and therefore the descision has been made to drop it too

```python
loan_df = loan_df.drop(['Unnamed: 0','id', 'total_rec_prncp'], axis=1)
```
